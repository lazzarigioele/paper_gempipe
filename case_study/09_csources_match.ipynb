# match with experimental data (specific)

```python
phe_spe = pnd.read_csv('reference/Lreuteri_phenotypes - specific.csv', index_col=0, skiprows=2)

# create columns for simulations: 
exp_cols = list(phe_spe.columns[2:])
sim_cols = [f"{acc}_sim" for acc in exp_cols]
for col in sim_cols: phe_spe[col] = None   # add columns as empty
all_cols = sorted(exp_cols + sim_cols)
phe_spe = phe_spe[list(phe_spe.columns[:2]) + all_cols]

print(phe_spe.shape)
phe_spe.head()
```

```python
# add sim values: 
cnps = pnd.read_csv('gempipe/output_derive/cnps.csv', index_col=0)

for acc in phe_spe.columns[2:]: 
    if acc.endswith('_sim') == True:
        continue 
    
    for index, row in phe_spe.iterrows(): 
        
        if acc not in cnps.columns:   # genome may be removd by quality-filtering:
            phe_spe.loc[index, f'{acc}_sim'] = '?'
        else:
            
            if row['exr_draft_pangsmm'] == '-': 
                phe_spe.loc[index, f'{acc}_sim'] = 0
            else:
                biggid = row['exr_draft_pangsmm'][3: -2]
                sim = cnps.loc[f'[C]{biggid}', acc]
                phe_spe.loc[index, f'{acc}_sim'] = sim 
    
print(phe_spe.shape)
sheet_1 = phe_spe.copy()
phe_spe
```

```python
# focus on xylose:
phe_spe[phe_spe['exr_draft_pangsmm']=='EX_xyl__D_e'].iloc[0, 2:]
```

```python
# compute match per strain:

accuracy_dict = {}

for acc in phe_spe.columns[2:]: 
    if acc.endswith('_sim'): 
        continue   # don't inspet same strain 2 times
       
    if list(phe_spe[f'{acc}_sim'].unique()) == ['?'] :  
        continue     # quality filterred genome!
        
    matches = 0
    for index in phe_spe.index: 
        
        exp_value = phe_spe.loc[index, acc]
        sim_value = phe_spe.loc[index, f'{acc}_sim']
            
        if exp_value == sim_value: 
            matches += 1
    
    accuracy_dict[acc] = round(matches /  len(phe_spe)*100, 0)
print(accuracy_dict)
```

```python
# get the mean accuracy: 

print(round(sum(accuracy_dict.values()) / len(accuracy_dict), 0))
```

# match with experimental data (general)

## using metabolic clusters

```python
# load clusterization from previous notebook:
with open("gempipe/acc_to_cluster.pkl", "rb") as file:
    acc_to_cluster = pickle.load(file)
```

```python
# define subspecies-to-cluster associations:
sub_to_cluster = {
    'Lr_rodentium': 5, 
    'Lr_suis': 2,
    'Lr_murium': 5, 
    'Lr_porcinus': 4,
    'Lr_kinnaridis': 4,
    'Lr_reuteri': 6,
}
```

```python
# add sim values: 
cnps = pnd.read_csv('gempipe/output_derive/cnps.csv', index_col=0)

for sub in phe_gen.columns[2:]: 
    if sub.endswith('_sim') == True:
        continue 
    
    # iterate tested substrates:
    for index, row in phe_gen.iterrows(): 
        
        
        # iterate accessions in this cluster: 
        cnt_feature_present = 0
        cnt_strains_in_cluster = 0
        
        for acc, cluster in acc_to_cluster.items():
            if sub_to_cluster[sub] != cluster: 
                continue
            
            cnt_strains_in_cluster += 1
                
            if row['exr_draft_pangsmm'] == '-': 
                cnt_feature_present += 0
            else:
                biggid = row['exr_draft_pangsmm'][3: -2]
                sim = cnps.loc[f'[C]{biggid}', acc]
                cnt_feature_present += sim
        
        
        # compute relative frequency
        rel_freq = cnt_feature_present / cnt_strains_in_cluster
        phe_gen.loc[index, f'{sub}_sim'] = rel_freq
        
    print(f'{sub} (Cluster_{sub_to_cluster[sub]}): {cnt_strains_in_cluster} strains')
    
print(phe_gen.shape)
sheet_2 = phe_gen.copy()
phe_gen
```

## using subspecies strictly

```python
report_updated = pnd.read_csv('gempipe/report_updated.csv', index_col = 0)
report_updated = report_updated.set_index('accession', drop=True)
report_updated
```

```python
# add sim values: 
cnps = pnd.read_csv('gempipe/output_derive/cnps.csv', index_col=0)

for sub in phe_gen.columns[2:]: 
    if sub.endswith('_sim') == True:
        continue 
    
    # iterate tested substrates:
    for index, row in phe_gen.iterrows(): 
        
        
        # iterate accessions in this subspecies: 
        cnt_feature_present = 0
        
        for acc in report_updated[report_updated['species']==sub].index:

            if row['exr_draft_pangsmm'] == '-': 
                cnt_feature_present += 0
            else:
                biggid = row['exr_draft_pangsmm'][3: -2]
                sim = cnps.loc[f'[C]{biggid}', acc]
                cnt_feature_present += sim
        
        
        # compute relative frequency
        rel_freq = cnt_feature_present / len(report_updated[report_updated['species']==sub])
        phe_gen.loc[index, f'{sub}_sim'] = rel_freq
        
    
print(phe_gen.shape)
sheet_3 = phe_gen.copy()
phe_gen
```
