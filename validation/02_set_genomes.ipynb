```python
import shutil
import glob

import pandas as pnd
```

# 01_klebsiella

```python
# prepare the 37 input genomes:
# mamba update openpyxl
metadata = pnd.read_csv('01_klebsiella/tables/metadata.csv', index_col=0)

metadata['full_name'] = metadata['Species'] + ' '+ metadata['Strain']
metadata['full_name'] = metadata['full_name'].str.replace('K. ', 'Klebsiella ')
metadata.index = metadata.index + 1
metadata = metadata.rename(columns={i: i.lower() for i in metadata.columns})
metadata[['full_name', 'accession']].to_excel('01_klebsiella/tables/genomes.xlsx')

print(metadata.shape)
metadata
```

```python
# create a folder containing the input genomes, 
# having as filename the strain ID instead of the assemlby accession.

for index, row in metadata.iterrows():
    # get the assembly file
    if row['accession'] == 'NC_022566.1 + NC_005249.1' : 
        assembly = '01_klebsiella/genomes/NC_022566.1_NC_005249.1.fasta'
    else: assembly = '01_klebsiella/genomes/' + row['accession'] + '.fna'
    # get the strain name
    species = row['species']
    if   species == 'K. africana': species = 'Ka'
    elif species == 'K. pneumoniae': species = 'Kp'
    elif species == 'K. quasipneumoniae subsp quasipneumoniae': species = 'Kqq'
    elif species == 'K. quasipneumoniae subsp similipneumoniae': species = 'Kqs'
    elif species == 'K. quasivariicola': species = 'Kqv'
    elif species == 'K. variicola subsp tropica': species = 'Kvt'
    elif species == 'K. variicola subsp variicola': species = 'Kvv'
    strain = str(row['strain'])
    if ' (=' in strain: strain = strain.split(' (=', 1)[0]
    strain  = strain.replace(' ', '').replace('-', '_')
    shutil.copyfile(assembly, f'01_klebsiella/genomes_renamed/{species}_{strain}.fna')
```

# 02_ralstonia

```python
metadata = pnd.read_csv('02_ralstonia/tables/metadata.csv', index_col=0)
metadata.index = metadata.reset_index(drop=True).index + 1
metadata = metadata.rename(columns={i: i.lower() for i in metadata.columns})

metadata['full_name'] = metadata['species'] + ' ' + metadata['strain']
metadata['full_name'] = metadata['full_name'].str.replace('R. ', 'Ralstonia ')

metadata[['full_name', 'accession']].to_excel('02_ralstonia/tables/genomes.xlsx')
metadata
```

```python
# create a folder containing the input genomes, 
# having as filename the strain ID instead of the assemlby accession.

for file in glob.glob('02_ralstonia/genomes/*fna'):
    accession = file.replace('02_ralstonia/genomes/', '')
    accession = '_'.join(accession.split('_')[:2])
    match = metadata[metadata['accession']==accession].iloc[0]
    renamed = match['strain'] + '.fna'
    shutil.copyfile(file, '02_ralstonia/genomes_renamed/' + renamed)
```

# 03_pseudomonas


```python
metadata = pnd.read_csv('03_pseudomonas/tables/metadata.csv', index_col=0)
metadata.index = metadata.reset_index(drop=True).index + 1
metadata = metadata.rename(columns={i: i.lower() for i in metadata.columns})
metadata = metadata.rename(columns={'phylogenetic subgroups': 'full_name'})
metadata[['full_name', 'accession']].to_excel('03_pseudomonas/tables/genomes.xlsx')
metadata
```

```python
# create a folder containing the input genomes, 
# having as filename the strain ID instead of the assemlby accession.

for file in glob.glob('03_pseudomonas/genomes/*fna'):
    accession = file.replace('03_pseudomonas/genomes/', '')
    accession = '_'.join(accession.split('_')[:2])
    match = metadata[metadata['accession']==accession].iloc[0]
    renamed = str(match['strain']) + '.fna'
    renamed = renamed.replace(' ', '_')
    shutil.copyfile(file, '03_pseudomonas/genomes_renamed/' + renamed)
```
